```python
n = int(0.8 * len(x))

x_train = x[:n]
y_train = y[:n]

x_test = x[n:]
y_test = y[n:]
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def hypothesis(a,b,x):
    theta = (a * x) + b
    return sigmoid(theta)
```

```python
def error(a,b,x,y):
    error = 0
    m = len(x)
    for i in range(m):
        error += (-y[i]*np.log(hypothesis(a,b,x[i]))) - ((1 - y[i])*np.log(1 - hypothesis(a,b,x[i])))
    return (1/m)*error
```

```python
def step_gradient(a,b,x,y,learning_rate):
    grad_a = 0
    grad_b = 0
    m = len(x)
    for i in range(m):
        grad_a += 2/m * (hypothesis(a,b,x[i]) - y[i]) * x[i]
        grad_b += 2/m * (hypothesis(a,b,x[i]) - y[i])
    
    a = a - (grad_a * learning_rate)
    b = b - (grad_b * learning_rate)
    
    return a, b
```

```python
def descend(initial_a, initial_b, x, y, learning_rate, iterations):
    a = initial_a
    b = initial_b
    for i in range(iterations):
        e = error(a, b, x ,y)
        if i % 1000 == 0:
            print(f"Error: {e}, a: {a}, b: {b}")
        
        a, b = step_gradient(a, b, x, y, learning_rate)

    return a, b
```

```python
a = 0
b = 1
learning_rate = 0.01
iterations = 10000
final_a, final_b = descend(a, b, x_train, y_train, learning_rate, iterations)
```
